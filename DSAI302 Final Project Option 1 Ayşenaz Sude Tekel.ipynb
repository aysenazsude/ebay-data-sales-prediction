### T-TEST for understanding the statistical relation between men and women parfumes

```python
from scipy import stats
from scipy.stats import ttest_ind

men_sales = df[df['sex'] == 'men']['sold']
womens_sales = df[df['sex'] == 'women']['sold']
t_stat, p_value = stats.ttest_ind(men_sales, womens_sales)
print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_value:.2e}")

if p_value < 0.05:
    print("There is a statistically significant difference between the groups.")
else:
    print("There is no statistically significant difference between the groups.")
```

## K-Means Cluster for Types and Prices

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


df['type'] = pd.Categorical(df['type'])
df['type_code'] = df['type'].cat.codes

type_mapping = dict(enumerate(df['type'].cat.categories))
print("Type Codes Table:")
for code, type_name in type_mapping.items():
    print(f"Type Code {code}: {type_name}")


X = df[['type_code']]


kmeans = KMeans(n_clusters=3)
kmeans.fit(X)


cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_

print("\nCluster Centers:")
print(cluster_centers)
print("\nLabels:")
print(labels)

df['cluster'] = labels

fig, axs = plt.subplots(1, len(cluster_centers), figsize=(18, 6), sharey=True)

for i in range(len(cluster_centers)):
    cluster_data = df[df['cluster'] == i]
    axs[i].scatter(cluster_data['type_code'], cluster_data['price'], label=f'Cluster {i+1}')
    axs[i].scatter(cluster_centers[i], cluster_data['price'].mean(), color='red', marker='x', label=f'Centroid {i+1}')
    axs[i].set_xlabel('Type Code')
    axs[i].set_ylabel('Price')
    axs[i].set_title(f'Cluster {i+1}')
    axs[i].legend()

for ax in axs:
    handles, labels = ax.get_legend_handles_labels()
    updated_labels = [f'{type_mapping[int(label.split(" ")[1])]} ({int(label.split(" ")[1])})' for label in labels]
    ax.legend(handles, updated_labels, loc='best')

plt.tight_layout()
plt.show()
```

**Analysis:** We see three graphs of products in different type codes. These graphs show that for some specific types, we can find items at various prices. For other types, the price variation is not as significant

## Chi-square test for brand and sex

```python
from scipy.stats import chi2_contingency

data = df[['brand', 'sex']]
contingency_table = pd.crosstab(data['brand'], data['sex'])

chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-square Value:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:")
expected
```

**Analysis:** Given the extremely small p-value, we reject the null hypothesis of independence. This means there is a statistically significant association between the sex and brands.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

all_features = ['price', 'available', 'type', 'brand']
df_model = pd.get_dummies(df[all_features])

target = df['sold']
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_model, target, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Model Selection

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score


models = [
    ("Linear Regression", LinearRegression()),
    ("Ridge Regression", Ridge()),
    ("Lasso Regression", Lasso()),
    ("Random Forest", RandomForestRegressor())
]


for name, model in models:
    scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='neg_mean_squared_error')
    avg_score = -scores.mean()
    print(f"{name} Cross-Validation MSE Score: {avg_score}")
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor

param_grid = {
    'n_estimators': [100, 150, 200],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 3, 4],
    'min_samples_leaf': [1, 2, 3]
}

random_search = RandomizedSearchCV(RandomForestRegressor(), param_grid, n_iter=50, cv=5, scoring='neg_mean_squared_error', n_jobs=-1, verbose=2)

random_search.fit(X_train_scaled, y_train)


print("Best Parameters:", random_search.best_params_)
print("Best Cross-Validation MSE Score:", -random_search.best_score_)
```

## Best model: Random Forest

```python
best_random_forest = RandomForestRegressor(**random_search.best_params_)
best_random_forest.fit(X_train_scaled, y_train)


y_train_pred = best_random_forest.predict(X_train_scaled)
train_mse = mean_squared_error(y_train, y_train_pred)
print("Training MSE Score:", train_mse)


y_test_pred = best_random_forest.predict(X_test_scaled)
test_mse = mean_squared_error(y_test, y_test_pred)
print("Test MSE Score:", test_mse)

```

```python
from scipy.stats import zscore
z_scores = zscore(y_test_pred)
outlier_mask = z_scores > 3

cleaned_y_test = y_test[~outlier_mask]
cleaned_y_test_pred = y_test_pred[~outlier_mask]

outlier_y_test = y_test[outlier_mask]
outlier_y_test_pred = y_test_pred[outlier_mask]

plt.figure(figsize=(10, 6))
plt.scatter(cleaned_y_test, cleaned_y_test_pred, color='green', alpha=0.5, label='Good Predictions')
plt.scatter(outlier_y_test, outlier_y_test_pred, color='red', alpha=0.5, label='Bad Predictions')

plt.xlabel("Actual Sold")
plt.ylabel("Predicted Sold")
plt.title("Actual vs. Predicted Sold (Random Forest)")
plt.legend()
plt.show()
```

**Analysis:** Our model provides accurate predictions for small quantities of sold items; however, it may yield inaccurate predictions for larger quantities.

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error

models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(),
    "Lasso Regression": Lasso()
}

param_grids = {
    "Linear Regression": {},
    "Ridge Regression": {'alpha': [0.1, 1, 10]},
    "Lasso Regression": {'alpha': [0.1, 1, 10]}
}


model_performances = {}
for name, model in models.items():
    print(f"Finding best parameters for {name}...")

    random_search = RandomizedSearchCV(model, param_grids[name], n_iter=50, cv=5, scoring='neg_mean_squared_error', n_jobs=-1, verbose=2)
    random_search.fit(X_train_scaled, y_train)

    print(f"Best Parameters for {name}:", random_search.best_params_)
    print(f"Best Cross-Validation MSE Score for {name}:", -random_search.best_score_)

    model_performances[name] = {
        'best_params': random_search.best_params_,
        'best_cv_mse': -random_search.best_score_
    }

    print()

for name, info in model_performances.items():
    print(f"Evaluating {name} with best parameters...")

    model = None
    if name == "Linear Regression":
        model = LinearRegression(**info['best_params'])
    elif name == "Ridge Regression":
        model = Ridge(**info['best_params'])
    elif name == "Lasso Regression":
        model = Lasso(**info['best_params'])

    model.fit(X_train_scaled, y_train)

    y_train_pred = model.predict(X_train_scaled)
    train_mse = mean_squared_error(y_train, y_train_pred)
    print(f"{name} Training MSE Score:", train_mse)

    y_test_pred = model.predict(X_test_scaled)
    test_mse = mean_squared_error(y_test, y_test_pred)
    print(f"{name} Test MSE Score:", test_mse)

    print()
```

```python
from scipy.stats import zscore
import matplotlib.pyplot as plt

y_test_preds = {}

for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_test_pred = model.predict(X_test_scaled)
    y_test_preds[name] = y_test_pred

for name, y_pred in y_test_preds.items():
    z_scores = zscore(y_pred)
    outlier_mask = z_scores > 3

    cleaned_y_test = y_test[~outlier_mask]
    cleaned_y_test_pred = y_pred[~outlier_mask]

    outlier_y_test = y_test[outlier_mask]
    outlier_y_test_pred = y_pred[outlier_mask]

    plt.figure(figsize=(10, 6))
    plt.scatter(cleaned_y_test, cleaned_y_test_pred, color='green', alpha=0.5, label='Good Predictions')
    plt.scatter(outlier_y_test, outlier_y_test_pred, color='red', alpha=0.5, label='Bad Predictions')

    plt.xlabel("Actual Sold")
    plt.ylabel("Predicted Sold")
    plt.title(f"Actual vs. Predicted Sold ({name})")
    plt.legend()
    plt.show()
```

**Analysis:** I chose Random Forest as the best model among these models because its performance was superior compared to Ridge Regression, Lasso Regression, and Linear Regression. The cross-validation MSE score of Random Forest was lower compared to the other models, indicating better predictive performance. Although Ridge Regression had a competitive MSE score, Random Forest outperformed it. Additionally, Random Forest is known for its robustness to overfitting and its ability to handle non-linear relationships in the data, which could have contributed to its superior performance in this case. Therefore, based on the evaluation of cross-validation MSE scores and considering the strengths of Random Forest, it was selected as the best model for this task.
